The means of the quantitative variables in the data set are calculated below.

```r
player_tibble <- tibble(players)


played_hours_mean <- player_tibble |>  # extracting the mean from the played_hours variable
    select(played_hours) 
    as.numeric() 
mean(pull(player_tibble, played_hours), na.rm = TRUE)   

Age_mean <- player_tibble |> # extracting the mean from the Age variable
    select(Age)  
    as.numeric() 
mean(pull(player_tibble, Age), na.rm = TRUE)
```

```r
summary_Age <- summary(Age_mean) # summary of Age, to prove mean 
summary_Age
```

```r

Age_plot <- players |> # plot of age of the individual compared to how many hours they play
    ggplot(aes(x = played_hours, y = Age)) +
    geom_point() +
    scale_x_continuous(limits = c(0,5)) +
    labs(x = "hours played", y = "Age") +
    ggtitle("Age of player compared to hours (up to 5 hours)") +
    theme(text = element_text(size=18)) 
Age_plot
```

This first plot is the age of the player compared to how many hours they played. In our question we are trying to discover the demographic of individuals that play more hours, this shows the trend that is it more common for younger players have more hours of playing. There were a few points (outliers) cut from this plot due to them being so large and few that it would not contribute to a trend.

```r
options(repr.plot.width = 13, repr.plot.height = 7) 
gender_plot <- players |> # plot of age of the individual compared to how many hours they play
    ggplot(aes(x = gender, y = played_hours, fill = gender)) +
    geom_bar(stat = "identity") +
    labs(x = "Gender", y = "Hours played") +
    ggtitle("Total hours played per gender") +
    theme(text = element_text(size=18)) 
gender_plot
```

This table shows the total hours played per gender, with male being the most. This will be helpful in our question as it again alludes to the demographic that spends the most hours on video games.
